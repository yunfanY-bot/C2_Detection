```python
import pandas as pd
import os
```

```python

# Current working directory
directory = '.'

# List all CSV files in the current directory
file_list = [file for file in os.listdir(directory) if file.startswith('dataflow_final_') and file.endswith('.csv')]

# Sort files to ensure correct order
file_list.sort()

# Read each file, skip the first row, and collect them into a list
dataframes = [pd.read_csv(filename) for filename in file_list]

# Concatenate all dataframes into one
integrated_df = pd.concat(dataframes, ignore_index=True)
print(integrated_df.columns)
# Write the integrated dataframe to a new CSV file
integrated_df.to_csv('dataflow_integrated.csv', index=False)
```

```python
print(integrated_df)
```

```python
# Getting the last column's name
last_column_name = integrated_df.columns[-1]

# Calculating value counts for the last column
value_counts = integrated_df[last_column_name].value_counts()

# Display the value counts
print(f"Value counts for the last column ('{last_column_name}'): \n{value_counts}")
```

```python
# Filter rows where the last column is 'normal'
normal_rows = integrated_df[integrated_df[last_column_name] == 'normal']

# Randomly sample 10% of the 'normal' rows
sampled_normal_rows = normal_rows.sample(frac=0.1, random_state=1)  # random_state for reproducibility

# Filter rows for 'merlin', 'dnscat', 'metasploit'
other_rows = integrated_df[integrated_df[last_column_name].isin(['merlin', 'dnscat', 'metasploit'])]

# Combine the sampled normal rows with the other specific rows
combined_rows = pd.concat([sampled_normal_rows, other_rows], ignore_index=True)

# Write the combined data to a new CSV file
combined_rows.to_csv('dataflow_balanced.csv', index=False)

print(combined_rows)
```

```python
# Getting the last column's name
last_column_name = combined_rows.columns[-1]

# Calculating value counts for the last column
value_counts = combined_rows[last_column_name].value_counts()

# Display the value counts
print(f"Value counts for the last column ('{last_column_name}'): \n{value_counts}")
```

```python
# Get the name of the last column
last_column_name = combined_rows.columns[-1]

# Calculate the mean for each column for each type in the last column, excluding the last column itself
averages = combined_rows.iloc[:, :-1].groupby(combined_rows[last_column_name]).mean()

# Print the column names and average values for each type
print("Average values by type for each column:")
for col_name in averages.columns:
    output = f"{col_name}:"
    for type_name in averages.index:
        avg_value = averages.at[type_name, col_name]
        output += f" {type_name}: {avg_value:.2f},"
    print(output.strip(','))
```
